# Planteamiento y esquema

Casca.png

Un cilindro hueco muy largo hecho de una material de permeabilidad magntica $\mu$ y de radios $R_a< R_b$ es colocada en el vacio en presencia de una campo magnetico $\vec{B}=B\hat{i}$ de manera que el eje del cilindro es perpendicular al campo.Obtenga los campos $\vec{B},\vec{H}$ en las distintas regiones, notese que dentro del cilindro tambin hay vacio.

# Desarrollo

No hay una corriente en el cilindro y tampoco en el espacio vacio.De la ley de Ampre para medios:
$$\vec{J}=0\quad \rightarrow \nabla\times H=0$$
Esto quiere decir por el teorema de Helmholtz y la ley de Gauss:
$$\vec{H}=-\nabla \varphi \qquad \nabla \cdot H=0 \rightarrow \nabla ^2 \varphi=0$$
El potencial escalar magntico cumple la ecuacin de Laplace. Tomaremos coordenadas cilndricas por lo que la solucin total es:
\begin{aligned}
\mathbb{V}_{0, \nu}(\rho, \theta, z)=\left(A_{0}\right.&\left.+B_{0} z\right)\left[\left(E_{0}+F_{0} \ln \frac{\rho}{\rho_{0}}\right)\left(C_{0}+D_{0} \theta\right)\right.\\
&\left.+\sum_{\nu = 0}\left(E_{\nu} \rho^{\nu}+F_{\nu} \rho^{-\nu}+\left.\int\left\{\left[E_{\nu} J_{\nu}(k \rho)+F_{\nu} N_{\nu}(k \rho)\right]\left(A(k) e^{k z}+B(k) e^{-k z}\right)\right\} d k\right\}\right)\left(C_{\nu} \cos \nu \theta+D_{\nu} \operatorname{sen} \nu \theta\right)\right]
\end{aligned}


##Consideraciones Fsicas
El cilindro que es muy largo se tomara como infinito, en este caso el potencial no depende de $z$ por lo que cualquier trmino que involucre a $z$ se eliminara, por lo tanto $A(k)=0,B(k)=0,B_0=0$.Adicionalmente vemos que visto en el plano $xy$ hay una simetria respecto al angulo $\theta$ de manera que aquellos que tengan un angulo $-\theta$ deben ser iguales a los que tengan un angulo $\theta$ debido a que importa el angulo que tengan respecto a $\vec{B}$ es decir que las funciones de $\theta$ que aparezcan deben ser pares. por esto $D_0=0,D_{\nu}=0$ Se obtiene:
$$
\varphi_{0, \nu}(\rho, \theta)=A_{0}+B_{0} \ln \frac{\rho}{\rho_{0}}+\sum_{\nu \neq 0}\left(A_{\nu} \rho^{\nu}+B_{\nu} \rho^{-\nu}\right) \cos \nu \theta
$$


Esta suma se igualara termino a termino con condiciones de continuidad por lo que se escribe :
\begin{array}{r}
\varphi_{0, \nu}(\rho, \theta)=A_{0}+B_{0} \ln \frac{\rho}{\rho_{0}}+A_{1} \rho \cos \theta+\frac{B_{1}}{\rho} \cos \theta \\
+A_{2} \rho^{2} \cos 2 \theta+\frac{B_{2}}{\rho^{2}} \cos 2 \theta+\cdots
\end{array}


Como es el potencial del campo $\overrightarrow{\mathcal{B}}$, esto lo utilizaremos para comparar con el potencial producido por el material:
\begin{aligned}
\overrightarrow{\mathcal{B}} &=-\mu_{0} \nabla \varphi_{\mathcal{B}} \\
\mathcal{B} \hat{\mathbf{i}} &=-\mu_{0}\left[\frac{\partial \varphi_{\mathcal{B}}}{\partial x} \hat{\mathbf{i}}+\frac{\partial \varphi_{\mathcal{B}}}{\partial y} \hat{\mathbf{j}}+\frac{\partial \varphi_{\mathcal{B}}}{\partial z} \hat{\mathbf{k}}\right] \\
-\frac{\mathcal{B}}{\mu_{0}} &=\frac{\partial \varphi_{\mathcal{B}}}{\partial x} \\
\varphi_{\mathcal{B}} &=-\frac{\mathcal{B}}{\mu_{0}} x+\varphi_{0}
\end{aligned}
Adems recordando que las coordenadas cilndricas son:
\begin{aligned}
x=\rho \cos(\theta) \quad y=\rho \sin(\theta) \quad z=z
\end{aligned}
El potencial del campo magntico $\overrightarrow{\mathcal{B}}$ esta dado en coordenadas cilindricas por:
$$\varphi_{\mathcal{B}}=-\frac{\mathcal{B}}{\mu_{0}} \rho \cos \theta+\varphi_{0}$$
Dividimos las regiones como $\varphi_f ,\varphi_d ,\varphi_c$ para fuera,dentro(regin en el vacio) y en el centro(donde hay material) del cilindro.Para el potencial $\varphi_f$ cuando se esta muy lejos de la influencia del potencial generado por el cilindro es de esperarse que permanezca la accin del campo $\overrightarrow{\mathcal{B}}$:
\begin{aligned}
\varphi_{\mathcal{B}}=\lim_{\rho \rightarrow \infty} \varphi_f\\
\lim_{\rho \rightarrow \infty} \varphi_f=\lim_{\rho \rightarrow \infty}A_{0}+B_{0} \ln \frac{\rho}{\rho_{0}}+A_{1} \rho \cos \theta+\frac{B_{1}}{\rho} \cos \theta \\
+A_{2} \rho^{2} \cos 2 \theta+\frac{B_{2}}{\rho^{2}} \cos 2 \theta+\cdots
\\=A_0+B_0\ln\frac{\rho}{\rho_0}+A_1\rho \cos\theta+A_2\rho^2\cos2\theta+\cdots
\end{aligned}
Se iguala con el potencial anteriormente obtenido
\begin{aligned}
-\frac{\mathcal{B}}{\mu_{0}}\rho \cos\theta+\varphi_{0}=A_0+B_0\ln\frac{\rho}{\rho_0}+A_1\rho \cos\theta+A_2\rho^2\cos2\theta+\cdots
\end{aligned}
De alli se deduce que :
\begin{array}{l}
A_{0}=\varphi_{0} \\
B_{0}=0 \\
A_{1}=-\frac{B}{\mu_{0}} \\
A_{\nu}=0, \quad \forall \nu \geqslant 2
\end{array}
De manera que tenemos hasta ahora que el potencial fuera del cilindro($\rho>R_b$) es:
$$\varphi_{f}(\rho, \theta)=\varphi_{0}-\frac{\mathcal{B}}{\mu_{0}} \rho \cos \theta+\frac{B_{1}}{\rho} \cos \theta+\frac{B_{2}}{\rho^{2}} \cos 2 \theta+\cdots$$
ahora el caso para el potencial dentro del cilindro ($\rho< R_a$):
\begin{aligned}
\varphi_{i}(\rho, \theta)=C_{0}+D_{0} \ln \frac{\rho}{\rho_{0}}+C_{1} \rho \cos \theta+\frac{D_{1}}{\rho} & \cos \theta \\
&+C_{2} \rho^{2} \cos 2 \theta+\frac{D_{2}}{\rho^{2}} \cos 2 \theta+\cdots
\end{aligned}
En este las potencias negativas de $\rho$no pueden aparecer ya que esta region incluye a $\rho=0$ , de manera que para que el potencial no diverja debe ser:
$$\varphi_{i}(\rho, \theta)=C_{0}+C_{1} \rho \cos \theta+C_{2} \rho^{2} \cos 2 \theta+\cdots$$
y en el centro el potencial es:
\begin{aligned}
\varphi_{c}(\rho, \theta)=E_{0}+F_{0} \ln \frac{\rho}{\rho_{0}}+E_{1} \rho \cos \theta+\frac{F_{1}}{\rho} & \cos \theta \\
&+E_{2} \rho^{2} \cos 2 \theta+\frac{F_{2}}{\rho^{2}} \cos 2 \theta+\cdots
\end{aligned}

$$\overrightarrow{\mathcal{B}}_{1} \cdot \hat{n}_{2}=\overrightarrow{\mathcal{B}}_{2} \cdot \hat{n}_{2} \quad \rightarrow \quad \overrightarrow{\mathcal{B}}_{i} \cdot \hat{\rho}=\overrightarrow{\mathcal{B}}_{c} \cdot \hat{\boldsymbol{\rho}}$$

\begin{aligned}

---

\begin{aligned}

Se escribe explicitamente el campo $\overrightarrow{\mathcal{B}}_i$, y se crean constantes auxiliares para visualizar el campo mejor

\begin{aligned}\overrightarrow{\mathcal{B}}_{i}=\frac{4 \mu \mu_{0} R_{b}^{2}}{\left(\mu+\mu_{0}\right)^{2} R_{b}^{2}-\mu_{0}^{2} \chi_{m}^{2} R_{a}^{2}} \overrightarrow{\mathcal{B}}=
\\
\frac{4 \mathcal{K}_m \mu_{0}^2 R_{b}^{2}}{\left(\mathcal{K}_m+1\right)^{2} \mu_0^2R_{b}^{2}-\mu_{0}^{2} \chi_{m}^{2} R_{a}^{2}} \mathcal{B}\hat{i}
\\
l=\frac{\mathcal{B}}{\left(\mathcal{K}_m+1\right)^{2} \mu_0^2R_{b}^{2}-\mu_{0}^{2} \chi_{m}^{2} R_{a}^{2}} 
\\
s=4 \mathcal{K}_m \mu_{0}^2 R_{b}^{2}
\\
\overrightarrow{\mathcal{B}}_{i}=s\cdot l\hat{i}
\end{aligned}

El campo $\overrightarrow{\mathcal{H}}_{i}$ es el siguiente:

$$\overrightarrow{\mathcal{H}}_{i}=\frac{\overrightarrow{\mathcal{B}}_{i}}{\mu_{0}}=\frac{1}{\mu} \frac{4 R_{b}^{2}}{\left(1+\frac{\mu_{0}}{\mu}\right)^{2} R_{b}^{2}-\left(1-\frac{\mu_{0}}{\mu}\right)^{2} R_{a}^{2}} \overrightarrow{\mathcal{B}}$$

Ahora seguimos con el campo $\overrightarrow{\mathcal{B}}_{f}$ y de igual manera se introducen unas constantes auxiliares

$$\overrightarrow{\mathcal{B}}_{f}=\overrightarrow{\mathcal{B}}+\left(\mu-\mu_{0}\right) \frac{R_{b}^{2}}{\rho^{2}} \frac{\mu_{0} R_{a}^{2}\left(\chi_{m}+2\right)-\left(\mu+\mu_{0}\right) R_{b}^{2}}{\mu_{0}^{2} \chi_{m}^{2} R_{a}^{2}-\left(\mu+\mu_{0}\right)^{2} R_{b}^{2}} \mathcal{B}(\cos \theta \hat{\boldsymbol{\rho}}+\operatorname{sen} \theta \hat{\boldsymbol{\theta}})
\\= \overrightarrow{\mathcal{B}} \hat{i}-\frac{\mu_0\chi_mR_b^2}{(x^2+y^2)}l\cdot\left((\chi_m+2)R_a^2\mu_0-\mu_0(\mathcal{K}_m+1)R_b^2\right)\left[\left(\frac{x^2-y^2}{x^2+y ^2}\right)\hat{i}+2\left(\frac{xy}{x^2+y^2}\right)\hat{j}\right]
\\o=\mu_0\chi_mR_b^2
\\ m=-l\cdot\left((\chi_m+2)R_a^2\mu_0-\mu_0(\mathcal{K}_m+1)R_b^2\right)
\\
\overrightarrow{\mathcal{B}}_{f}=\mathcal{B}\hat{i}+\frac{o\cdot m \cdot }{(x^2+y^2)}\left[\left(\frac{x^2-y^2}{x^2+y ^2}\right)\hat{i}+2\left(\frac{xy}{x^2+y^2}\right)\hat{j}\right]
$$

Y el campo $\overrightarrow{\mathcal{H}}_{f}$:
\begin{array}{l}
\overrightarrow{\mathcal{H}}_{f}=\frac{\overrightarrow{\mathcal{B}}_{f}}{\mu_{0}} \\
\overrightarrow{\mathcal{H}}_{f}=\frac{\overrightarrow{\mathcal{B}}}{\mu_{0}}+\left(\mu-\mu_{0}\right) \frac{R_{b}^{2}}{\rho^{2}} \frac{\mu_{0} R_{a}^{2}\left(\chi_{m}+2\right)-\left(\mu+\mu_{0}\right) R_{b}^{2}}{\mu_{0}^{3} \chi_{m}^{2} R_{a}^{2}-\mu_{0}\left(\mu+\mu_{0}\right)^{2} R_{b}^{2}} \mathcal{B}(\cos \theta \hat{\boldsymbol{\rho}}+\operatorname{sen} \theta \hat{\boldsymbol{\theta}})
\end{array}

Finalmente el campo en el material:

\begin{aligned}
\overrightarrow{\mathcal{B}}_{c}=\frac{2 \mu\left(\mu+\mu_{0}\right) R_{b}^{2}}{\left(\mu+\mu_{0}\right)^{2} R_{b}^{2}-\left(\mu-\mu_{0}\right)^{2} R_{a}^{2}} \mathcal{B}(\cos \theta \hat{\rho}-\operatorname{sen} \theta \hat{\boldsymbol{\theta}}) & \\
-\frac{R_{a}^{2}}{\rho^{2}} \frac{2 \mu\left(\mu-\mu_{0}\right) R_{b}^{2}}{\left(\mu+\mu_{0}\right)^{2} R_{b}^{2}-\left(\mu-\mu_{0}\right)^{2} R_{a}^{2}} \mathcal{B}(\cos \theta \hat{\boldsymbol{\rho}}+\operatorname{sen} \theta \hat{\boldsymbol{\theta}})

n=2\mathcal{K}_m \mu_0^2(\mathcal{K}_m+1)R_b^2l\\

\overrightarrow{\mathcal{B}}_{c}=n\hat{i}-\frac{p}{(x^2+y^2)}\left[\left(\frac{x^2-y^2}{x^2+y ^2}\right)\hat{i}+2\left(\frac{xy}{x^2+y^2}\right)\hat{j}\right]

```
#Constantes
mu0=4*math.pi*10**(-7) # T*m*A-1 o H/M o N*A-2
xm=5500#mu-mu0/mu0
km=5500# mu/mu0
Ra=1#metros
Rb=3#metros
B=1#T teslas 
l=B/((km+1)**2*mu0**2*Rb**2-(km-1)**2*mu0**2*Ra**2)
s=4*km*mu0**2*Rb**2
m=-(Ra**2*(xm+2)*mu0-(km+1)*Rb**2*mu0)*l
o=xm*mu0*Rb**2
n=2*km*mu0**2*(km+1)*Rb**2*l
p=Ra**2*2*xm*mu0**2*(km-1)*Rb**2*l
```

```
U,V,X,Y=[],[],[],[]
rang=np.linspace(-Ra,Ra,30)
for x in rang:
  for y in rang:
    rho2=x**2+y**2
    if rho2<=Ra*Ra:
      U.append(s*l)
      V.append(0)
      X.append(x)
      Y.append(y)
      

U1,V1,X1,Y1=[],[],[],[]
rang=np.linspace(-Rb-5,Rb+5,40)
for x1 in rang:
  for y1 in rang:
    rho2=x1**2+y1**2
    if rho2 >=Rb*Rb :
      resta=(x1/math.sqrt(rho2))**2-(y1/math.sqrt(rho2))**2
      suma=(x1/math.sqrt(rho2))*(y1/math.sqrt(rho2))
      U1.append(B+(o*m*resta/rho2))
      V1.append(o*(m/rho2)*(2*suma))
      X1.append(x1)
      Y1.append(y1)
U2,V2,X2,Y2=[],[],[],[]
rang=np.linspace(-Rb,Rb,40)
for x2 in rang:
  for y2 in rang:
    rho2=x2**2+y2**2
    if  Ra*Ra<=rho2 and Rb*Rb>=rho2 :
      resta=(x2/math.sqrt(rho2))**2-(y2/math.sqrt(rho2))**2
      suma=(x2/math.sqrt(rho2))*(y2/math.sqrt(rho2))
      U2.append(n-(p*resta/rho2))
      V2.append(-2*p*(suma/rho2))
      X2.append(x2)
      Y2.append(y2)

figure = plt.figure()
#plt.figure(figsize=(10,10))
draw_circle =plt.Circle((0, 0), Ra,fill=False)
axes = figure.add_axes([7, 7, 6,7])
draw_circle2 =plt.Circle((0, 0), Rb,fill=False)
axes.add_artist(draw_circle)
axes.add_artist(draw_circle2)
w=0.0009
quiv=plt.quiver(X,Y,U,V,width=w,color='red',scale=75)
quiv1=plt.quiver(X1,Y1,U1,V1, width=w,scale=75)
quiv2=plt.quiver(X2,Y2,U2,V2,width=w,color='blue',scale=75)
plt.rcParams["font.size"] =80
plt.xlabel('x')
plt.ylabel('y')
axes.xaxis.zoom(0)
axes.yaxis.zoom(0)
print(U)
print(U1)
print(U2)
```
